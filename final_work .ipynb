## Normalizing continues values

### Normalize all temperatures together

We chose to normalize the columns
'Maximum Upper Bloom Temperature' </br>
'Minimum Upper Bloom Temperature' </br>
'Average Upper Bloom Temperature' </br>
'Average Lower Bloom Temperature'</br>
'Maximum Lower Bloom Temperature' </br>
'Minimum Lower Bloom Temperature' </br>
together to maintain correct proportions between the data, since Upper should be the largest of the data, Average should be the middle, and Lower should be the lowest. We will use max-min normalization to make them to be from range 0 to 1.

```python
all_temperature = pd.concat([data['Maximum Upper Bloom Temperature'], data['Minimum Upper Bloom Temperature'],
                             data['Average Upper Bloom Temperature'], data['Average Lower Bloom Temperature'],
                             data['Maximum Lower Bloom Temperature'], data['Minimum Lower Bloom Temperature']], ignore_index=True)
max_all_temperature = all_temperature.max()
min_all_temperature = all_temperature.min()
#Normalization of all temperature
data['Maximum Upper Bloom Temperature'] = (data['Maximum Upper Bloom Temperature']-min_all_temperature)/(max_all_temperature-min_all_temperature)
data['Minimum Upper Bloom Temperature'] = (data['Minimum Upper Bloom Temperature']-min_all_temperature)/(max_all_temperature-min_all_temperature)
data['Average Upper Bloom Temperature'] = (data['Average Upper Bloom Temperature']-min_all_temperature)/(max_all_temperature-min_all_temperature)
data['Average Lower Bloom Temperature'] = (data['Average Lower Bloom Temperature']-min_all_temperature)/(max_all_temperature-min_all_temperature)
data['Maximum Lower Bloom Temperature'] = (data['Maximum Lower Bloom Temperature']-min_all_temperature)/(max_all_temperature-min_all_temperature)
data['Minimum Lower Bloom Temperature'] = (data['Minimum Lower Bloom Temperature']-min_all_temperature)/(max_all_temperature-min_all_temperature)
```
